```python
import json

TABLE_NAME = "product_transaction_history"

# Schema dạng dictionary để dễ convert sang JSON
SCHEMA = {
    "fields": [
        {"name": "TransactionID", "type": "IntegerType", "nullable": True},
        {"name": "ProductID", "type": "IntegerType", "nullable": True},
        {"name": "ReferenceOrderID", "type": "IntegerType", "nullable": True},
        {"name": "ReferenceOrderLineID", "type": "IntegerType", "nullable": True},
        {"name": "TransactionDate", "type": "TimestampType", "nullable": True},
        {"name": "TransactionType", "type": "StringType", "nullable": True},
        {"name": "Quantity", "type": "IntegerType", "nullable": True},
        {"name": "ActualCost", "type": "DecimalType(19,4)", "nullable": True},
        {"name": "ModifiedDate", "type": "TimestampType", "nullable": True}
    ]
}

PRIMARY_KEYS = "TransactionID,ProductID"
```

```python
print(f"Bắt đầu incremental load bảng: {TABLE_NAME}")

schema_json = json.dumps(SCHEMA)

result = mssparkutils.notebook.run(
    path = "Get Data From Supabase Incremental load",
    timeout_seconds = 300,
    arguments = {
        "table_name": TABLE_NAME,
        "schema_json": schema_json,
        "primary_keys": PRIMARY_KEYS
    }
)

print(f"\nKết quả: {result}")

# Chỉ check nếu có lỗi thật sự (không chứa SUCCESS)
if "SUCCESS" not in result and result.startswith("FAILED:"):
    print(f"Load thất bại!")
    raise Exception(f"Load failed: {result}")
else:
    print(f"Load hoàn tất bảng {TABLE_NAME}!")
```
