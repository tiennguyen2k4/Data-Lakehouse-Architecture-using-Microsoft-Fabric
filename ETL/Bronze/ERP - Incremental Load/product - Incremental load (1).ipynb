```python
import json

TABLE_NAME = "product"

# Schema dạng dictionary để dễ convert sang JSON
SCHEMA = {
    "fields": [
        {"name": "ProductID", "type": "IntegerType", "nullable": True},
        {"name": "Name", "type": "StringType", "nullable": True},
        {"name": "ProductNumber", "type": "StringType", "nullable": True},
        {"name": "MakeFlag", "type": "StringType", "nullable": True},
        {"name": "FinishedGoodsFlag", "type": "StringType", "nullable": True},
        {"name": "Color", "type": "StringType", "nullable": True},
        {"name": "SafetyStockLevel", "type": "IntegerType", "nullable": True},
        {"name": "ReorderPoint", "type": "IntergerType", "nullable": True},
        {"name": "StandardCost", "type": "DecimalType(19,4)", "nullable": True},
        {"name": "ListPrice", "type": "DecimalType(19,4)", "nullable": True},
        {"name": "Size", "type": "StringType", "nullable": True},
        {"name": "SizeUnitMeasureCode", "type": "StringType", "nullable": True},
        {"name": "WeightUnitMeasureCode", "type": "StringType", "nullable": True},
        {"name": "Weight", "type": "DecimalType(8,2)", "nullable": True},
        {"name": "DaysToManufacture", "type": "IntegerType", "nullable": True},
        {"name": "ProductLine", "type": "StringType", "nullable": True},
        {"name": "Class", "type": "StringType", "nullable": True},
        {"name": "Style", "type": "StringType", "nullable": True},
        {"name": "ProductSubcategoryID", "type": "IntegerType", "nullable": True},
        {"name": "ProductModelID", "type": "IntegerType", "nullable": True},
        {"name": "SellStartDate", "type": "TimestampType", "nullable": True},
        {"name": "SellEndDate", "type": "TimestampType", "nullable": True},
        {"name": "rowguid", "type": "StringType", "nullable": True},
        {"name": "ModifiedDate", "type": "TimestampType", "nullable": True}
    ]
}

PRIMARY_KEYS = "ProductID"
```

```python
print(f"Bắt đầu incremental load bảng: {TABLE_NAME}")

schema_json = json.dumps(SCHEMA)

result = mssparkutils.notebook.run(
    path = "Get Data From Supabase Incremental load",
    timeout_seconds = 300,
    arguments = {
        "table_name": TABLE_NAME,
        "schema_json": schema_json,
        "primary_keys": PRIMARY_KEYS
    }
)

print(f"\nKết quả: {result}")

# Chỉ check nếu có lỗi thật sự (không chứa SUCCESS)
if "SUCCESS" not in result and result.startswith("FAILED:"):
    print(f"Load thất bại!")
    raise Exception(f"Load failed: {result}")
else:
    print(f"Load hoàn tất bảng {TABLE_NAME}!")
```
