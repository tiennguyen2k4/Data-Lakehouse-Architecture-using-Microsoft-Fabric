```python
import json

TABLE_NAME = "transaction"

# Schema dạng dictionary để dễ convert sang JSON
SCHEMA = {
    "fields": [
        {"name": "transaction_id", "type": "IntegerType", "nullable": True},
        {"name": "transaction_code", "type": "StringType", "nullable": True},
        {"name": "transaction_line_id", "type": "IntegerType", "nullable": True},
        {"name": "account_id", "type": "IntegerType", "nullable": True},
        {"name": "account_code", "type": "IntegerType", "nullable": True},
        {"name": "class_id", "type": "IntegerType", "nullable": True},
        {"name": "department_id", "type": "IntegerType", "nullable": True},
        {"name": "department_code", "type": "StringType", "nullable": True},
        {"name": "transaction_type", "type": "StringType", "nullable": True},
        {"name": "entry_type", "type": "StringType", "nullable": True},
        {"name": "description", "type": "StringType", "nullable": True},
        {"name": "transaction_date", "type": "TimestampType", "nullable": True},
        {"name": "debit_amount", "type": "DecimalType(19,4)", "nullable": True},
        {"name": "credit_amount", "type": "DecimalType(19,4)", "nullable": True},
        {"name": "amount", "type": "DecimalType(19,4)", "nullable": True},
        {"name": "ModifiedDate", "type": "TimestampType", "nullable": True}
    ]
}
```

```python
print(f"Bắt đầu load bảng: {TABLE_NAME}")

schema_json = json.dumps(SCHEMA)

result = mssparkutils.notebook.run(
    path = "Get Data From Supabase",
    timeout_seconds = 300,
    arguments = {
        "table_name": TABLE_NAME,
        "schema_json": schema_json,
        "drop_before_load": "true"
    }
)

print(f"\nKết quả: {result}")

# Chỉ check nếu có lỗi thật sự (không chứa SUCCESS)
if "SUCCESS" not in result and result.startswith("FAILED:"):
    print(f"Load thất bại!")
    raise Exception(f"Load failed: {result}")
else:
    print(f"Load hoàn tất bảng {TABLE_NAME}!")
```
